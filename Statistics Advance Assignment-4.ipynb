# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

```python
The t-test and z-test are both statistical tests used to compare means, but they differ in their assumptions and
applications.

1.Z-Test
-->Assumptions:
The sample size is large (typically n>30).
The population variance is known.
The data follows a normal distribution.

-->When to use a Z-test:
When you have a large sample size and the population variance is known.
Example Scenario: You want to test whether the average height of a population of adult males is different from a
known population mean (e.g., 5 feet 10 inches), and you know the population standard deviation.


2.T-Test
-->Assumptions:
The sample size is small (typically n<30).
The population variance is unknown (you estimate it from the sample).
The data follows a normal distribution.

-->When to use a T-test:
When you have a small sample size and the population variance is unknown.
Example Scenario: You want to test whether a new teaching method affects the test scores of students. You have a
small sample of students (e.g., 15 students), and you don't know the population variance.


Example Scenarios:
1.Z-Test Example: You are studying the average SAT score of a large group of students. The national average SAT
score is 1050, with a known standard deviation of 200. You have a sample of 100 students and want to test if their
average SAT score is different from the national average.

2.T-Test Example: A pharmaceutical company tests a new drug on a small group of 20 patients to see if it lowers
blood pressure more effectively than the current standard treatment. Since the sample size is small and the
population variance is unknown, a t-test would be appropriate.
```

# Q2: Differentiate between one-tailed and two-tailed tests.

```python
1.Directionality:

One-tailed tests focus on deviations in one direction (greater than or less than).
Two-tailed tests look for deviations in both directions (not equal to).

2.Critical Value Placement:

In a one-tailed test, the critical value is located in one tail of the distribution (either left or right).
In a two-tailed test, the critical values are located in both tails of the distribution, with half of the
significance level (α) allocated to each tail.

3.Usage:

One-tailed tests are used when you have a specific hypothesis about the direction of the effect.
Two-tailed tests are used when you are testing for any significant difference, regardless of the direction.

4.Example:

One-tailed test:- Suppose a manufacturer claims that their lightbulbs last at least 1000 hours. You are only
interested in testing whether the lifespan of their lightbulbs is less than 1000 hours (left-tailed test).
Hypotheses:
H₀: The average lifespan of the lightbulbs is ≥ 1000 hours.
H₁: The average lifespan of the lightbulbs is < 1000 hours.

Two-tailed test:- Suppose you want to test whether the average weight of a certain breed of dog is different from 30
pounds. You are interested in detecting any difference, whether the weight is more than or less than 30 pounds.
Hypotheses:
H₀: The average weight of the dogs is 30 pounds.
H₁: The average weight of the dogs is not equal to 30 pounds.
```

# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

```python
In hypothesis testing, Type I and Type II errors are the two types of errors that can occur when making decisions
about the null hypothesis

1.Type I Error
Definition: A Type I error occurs when the null hypothesis (H₀) is incorrectly rejected when it is actually true.
In other words, it is a false positive result.

Significance Level (α): The probability of making a Type I error is denoted by the significance level, α, typically
set at 0.05 or 5%. This means that there is a 5% risk of rejecting a true null hypothesis.

Example Scenario: Suppose a pharmaceutical company tests a new drug and concludes that the drug is effective when,
in reality, it is not. Here, they have made a Type I error by rejecting the null hypothesis (which stated that the
drug is not effective) when it was actually true.

Consequences: A Type I error could lead to false conclusions, such as approving a drug that doesn't work or assuming
that a treatment is beneficial when it is not.


2.Type II Error
Definition: A Type II error occurs when the null hypothesis (H₀) is not rejected when it is actually false. In other
words, it is a false negative result.

Example Scenario: Suppose a medical researcher is testing whether a new drug lowers blood pressure, and they fail to
detect the drug's true effect, concluding that the drug does not work when it actually does. Here, they have made a
Type II error by failing to reject a false null hypothesis (which stated that the drug is ineffective).

Consequences: A Type II error could result in missed opportunities, such as not adopting an effective treatment or
not recognizing a significant effect.
```

# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

```python
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true
population parameter (such as the mean or proportion) with a certain level of confidence. It provides an estimated
range of values which is believed to contain the parameter with a specific probability, known as the confidence
level.

Steps to Calculate Confidence Interval:
1. Determine the sample statistic (e.g., the sample mean).
2. Select a confidence level (e.g., 95% confidence level).
3. Find the corresponding Z-value or t-value for the selected confidence level.
4. Calculate the margin of error using the standard deviation and sample size.
5. Construct the confidence interval by adding and subtracting the margin of error from the sample statistic.

Example of Confidence Interval:

For example, if you are estimating a 95% confidence interval around the mean proportion of female babies born every
year based on a random sample of babies, you might find an upper bound of 0.56 and a lower bound of 0.48. These are
the upper and lower bounds of the confidence interval. The confidence level is 95%.
```

# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

```python
Sample Problem
Problem Statement:
A medical test is used to detect a particular disease. The probability that a person has the disease is 0.01 (1%).
The test is 99% accurate, meaning that:

If a person has the disease, the probability of testing positive is 0.99 (true positive rate).
If a person does not have the disease, the probability of testing negative is 0.99 (true negative rate).
If a person tests positive, what is the probability that they actually have the disease?

Solution:
Define the Events:
A: The event that a person has the disease.
B: The event that a person tests positive for the disease.
Identify the Given Probabilities:

Prior probability of having the disease, P(A): 0.01
Probability of testing positive given the disease, P(B∣A): 0.99
Probability of testing negative given no disease, P(Negative∣No Disease): 0.99
Therefore, the probability of testing positive given no disease, P(B∣No Disease): 1 - 0.99 = 0.01


Calculate the Total Probability of Testing Positive, P(B):

𝑃(𝐵)=𝑃(𝐵∣𝐴) * 𝑃(𝐴) + 𝑃(𝐵∣No Disease) * 𝑃(No Disease)
P(B)=P(B∣A)⋅P(A)+P(B∣No Disease)⋅P(No Disease)
𝑃(𝐵) = (0.99 * 0.01) + (0.01 * 0.99)
P(B) = 0.0099 + 0.0099 = 0.0198


Apply Bayes' Theorem: P(A/B) = P(B/A).P(A)/P(B)
                      P(A/B) = (0.99 * 0.01)/0.0198
                      P(A/B) = 0.0099/0.0198 ≈0.5

Conclusion
If a person tests positive for the disease, the probability that they actually have the disease is approximately 50%.
Despite the high accuracy of the test, because the disease is rare, the probability of having the disease given a
positive test result is not as high as one might intuitively expect.
```

# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

```python
-->What is the Margin of Error in a Confidence Interval?

The margin of error (MOE) in a confidence interval is the range above and below the sample statistic (such as the
sample mean) that accounts for the uncertainty in the estimate of the population parameter. It reflects the level
of precision in the estimate and is influenced by factors such as the sample size, variability in the data, and the
confidence level.

-->How Does Sample Size Affect the Margin of Error?

The margin of error is inversely proportional to the square root of the sample size. This means that as the sample
size increases, the margin of error decreases. A larger sample size provides more information, leading to a more
precise estimate of the population parameter and hence a smaller margin of error.

Small Sample Size: Larger margin of error, leading to wider confidence intervals.
Large Sample Size: Smaller margin of error, leading to narrower confidence intervals.


Example Scenario:
Imagine you're conducting a survey to estimate the average amount of time people spend exercising per week in your
city. You initially take a small sample of 25 people, and the sample mean is 5 hours per week, with a standard
deviation of 2 hours. You calculate a 95% confidence interval based on this data.

Scenario 1: Small Sample Size (n = 25)
Sample Size n=25
Sample Mean =5 hours
Standard Deviation σ=2 hours
Z-value for 95% Confidence Level = 1.96

Margin of error = 1.96 * (2/5) = 0.784

The confidence interval is: CI=5±0.784=[4.216,5.784]


Scenario 2: Larger Sample Size (n = 100)
Now, suppose you conduct the survey with a larger sample of 100 people, but the sample mean and standard deviation remain the same.

Sample Size n=100
Sample Mean =5 hours
Standard Deviation σ=2 hours
Z-value for 95% Confidence Level = 1.96

The margin of error is 1.96 * (2/10) = 0.392

The confidence interval is: CI=5±0.392=[4.608,5.392]


Conclusion:
With a sample size of 25: The margin of error is 0.784, leading to a confidence interval of [4.216, 5.784].
With a sample size of 100: The margin of error decreases to 0.392, leading to a narrower confidence interval of
[4.608, 5.392].
```

# Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

```python
To calculate the z-score for a data point, use the formula:

𝑧 = X−μ/σ

Where:
X is the data point (75 in this case),
μ is the population mean (70),
σ is the population standard deviation (5).

Step 1: Plug the values into the formula: z = X−μ/σ, z = 75-70/5 = 5/5 = 1

Step 2: Interpretation of the z-score:
A z-score of 1 means that the data point (75) is 1 standard deviation above the population mean (70).

Positive z-score: The data point is above the mean.
Negative z-score: The data point would be below the mean.

Conclusion:
The z-score of 1 indicates that the value of 75 is 1 standard deviation above the mean. This suggests that the data
point is relatively close to the mean, but slightly higher than average compared to the rest of the data in the
population.
```

# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

```python
To determine if the new weight loss drug is significantly effective, we will conduct a one-sample t-test. This test
will compare the sample mean weight loss to a hypothesized population mean (e.g., no weight loss, which would mean 0
pounds). We will test whether the average weight loss of 6 pounds in the sample is significantly different from 0.

Step 1: Define the Hypotheses
Null Hypothesis (H₀): The drug has no effect on weight loss. The mean weight loss is 0 pounds.
                      H0:μ=0
Alternative Hypothesis (H₁): The drug is effective, and the mean weight loss is greater than 0 pounds.
                      H1:μ>0
This is a one-tailed t-test because we are specifically testing if the weight loss is greater than zero.


Step 2: Collect and Define Sample Data
Sample mean =6 pounds
Sample standard deviation s=2.5 pounds
Sample size n=50
Significance level α=0.05
Degrees of freedom df=n−1=50−1=49


Step 3: Calculate the Test Statistic (t-value)
The t-value is calculated using the formula: t = 6-0/(2.5/7.071) = 6/0.3536 = 16.98


Step 4: Determine the Critical Value
For a one-tailed t-test at a 95% confidence level and 49 degrees of freedom, we need to find the critical t-value
from the t-distribution table.

From the t-distribution table (or using statistical software), the critical value for a one-tailed test with 
df=49 at α=0.05 is approximately 1.676.


Step 5: Compare the Test Statistic to the Critical Value
Calculated t-value: 16.98
Critical t-value: 1.676
Since the calculated t-value (16.98) is much larger than the critical t-value (1.676), we reject the null
hypothesis.


Step 6: Conclusion
At the 95% confidence level, there is strong evidence to suggest that the drug is significantly effective in causing
weight loss. The average weight loss of 6 pounds observed in the sample is significantly greater than 0 pounds.
Therefore, the new weight loss drug is effective.
```

# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
```python
To determine if there is a significant difference in student performance between the two teaching methods, we will
conduct a two-sample t-test. This test compares the means of two independent samples to determine if there is a
statistically significant difference between them.

Step 1: Define the Hypotheses
Null Hypothesis (H₀): The two teaching methods have no effect on student performance. The means of the two groups
are equal.

𝐻0:𝜇𝐴=𝜇𝐵
 
Alternative Hypothesis (H₁): The two teaching methods have different effects on student performance. The means of
the two groups are not equal.

𝐻1:𝜇𝐴≠𝜇𝐵

This is a two-tailed t-test because we are testing for any significant difference, regardless of direction.


Step 2: Collect and Define Sample Data
Sample A:

Mean =85
Standard deviation = 6
Sample size (needs to be provided or assumed; if not provided, let's assume nA=30)

Sample B:

Mean =82
Standard deviation =5
Sample size nB(needs to be provided or assumed; if not provided, let's assume nB=30)


Step 3: Calculate the Test Statistic (t-value)
SEdiff ≈1.428
Next, calculate the t-value:
t = 85-82/1.428 ≈ 3/1.428 ≈ 2.10

Step 4: Determine the Degrees of Freedom
df ≈ 4.13/0.089 ≈ 46.4

For practical purposes, we round down to the nearest whole number, so df = 46.


Step 5: Find the Critical t-value
For a two-tailed test with α=0.01 and df≈46, use the t-distribution table or statistical software to find the
critical value. The critical t-value for 46 degrees of freedom at α=0.01 (two-tailed) is approximately 2.68.


Step 6: Compare the Test Statistic to the Critical Value
Calculated t-value: 2.10
Critical t-value: 2.68
Since the calculated t-value (2.10) is less than the critical t-value (2.68), we fail to reject the null hypothesis.

Conclusion
At the 0.01 significance level, there is not enough evidence to suggest a significant difference in student 
erformance between the two teaching methods.
```

# Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

```python
To determine if caffeine has a significant effect on reaction time, we'll conduct a one-sample t-test. This test
will compare the sample mean reaction time to a hypothesized population mean. For this test, we'll assume the null
hypothesis is that caffeine does not affect reaction time, meaning the average reaction time with caffeine is the
same as the population mean reaction time without caffeine.

Let's assume the hypothesized population mean reaction time (without caffeine) is 0.30 seconds for this test.

Step 1: Define the Hypotheses
Null Hypothesis (H₀): Caffeine does not affect reaction time. The mean reaction time with caffeine is 0.30 seconds.
                        H0:μ=0.30
Alternative Hypothesis (H₁): Caffeine affects reaction time. The mean reaction time with caffeine is different from
0.30 seconds.
                        𝐻1:𝜇≠0.30

This is a two-tailed t-test because we are testing for any significant difference, regardless of direction.


Step 2: Collect and Define Sample Data
Sample Mean : 0.25 seconds
Sample Standard Deviation : 0.05 seconds
Sample Size n: 30
Significance Level α: 0.10 (for a 90% confidence level)
Degrees of Freedom n−1=30−1=29


Step 3: Calculate the Test Statistic (t-value)
t = 0.25-0.30/0.0091 ≈-0.05/0.0091 ≈−5.49


Step 4: Determine the Critical t-value
For a two-tailed t-test with α=0.10 and df=29, the critical t-value can be found from the t-distribution table or
using statistical software. The critical t-value for a 90% confidence level and 29 degrees of freedom is
approximately ±1.699.


Step 5: Compare the Test Statistic to the Critical Value
Calculated t-value: -5.49
Critical t-value: ±1.699
Since the calculated t-value (-5.49) is less than the negative critical t-value (-1.699), we reject the null
hypothesis.

Conclusion
At the 90% confidence level, there is strong evidence to suggest that caffeine has a significant effect on reaction
time. The mean reaction time with caffeine is significantly different from the hypothesized population mean of 0.30
seconds.
```
